Establishing connection with duckdb

```python
con = duckdb.connect()
```

Registering the sales_table to duckdb

```python
con.register('sales_table', df)
```

```python
result=con.sql("SELECT * FROM sales_table LIMIT 5").fetchdf()
result
```

Observation-1: Converting the date column to a month and year format

```python
result = con.sql("""
    SELECT 
        EXTRACT(YEAR FROM CAST(date AS DATE)) AS Year,
        EXTRACT(MONTH FROM CAST(date AS DATE)) AS Month,Date,
        product_id,
        quantity_sold
    FROM sales_table
    ORDER BY Year,Month,Product_id,quantity_sold ASC
""").fetchdf()
result
```

Registering the updated_sales_table in task-1 to duckkdb 

```python
con.register('updated_sales_table', result)
```

Observation-2 Grouping the sales data by month,year and product_id

```python
sol=con.sql("""
                SELECT Year,Month,Product_id,SUM(quantity_sold)
                FROM updated_sales_table
                GROUP BY 
                Year,Month,Product_id
                ORDER BY Year ASC
            """).fetchdf()
sol
```

Observation-3 Calculating the average daily sales quantity for each product in each month

```python
sol=con.sql("""
    SELECT 
        product_id,
       Month,
        SUM(quantity_sold) / COUNT(DISTINCT CAST(date AS DATE)) AS avg_daily_sales_quantity
    FROM updated_sales_table
    GROUP BY product_id,Month
    ORDER BY product_id,Month ASC
""").fetchdf()
sol
```

Registering the above avg_daily_sales table as avg_sales to duckdb

```python
con.register('avg_sales', sol)
```

Observation-4 Identifying the month with highest average daily sales quantity for each product

```python
solution = con.sql("""
    WITH  max_avg_sales AS (
        SELECT product_id,
               MAX(avg_daily_sales_quantity) AS max_avg_daily_sales_quantity
        FROM avg_sales
        GROUP BY product_id
    )
    SELECT s.product_id,
           s.Month,
           m.max_avg_daily_sales_quantity
    FROM avg_sales s
    JOIN max_avg_sales m
    ON s.product_id = m.product_id
    AND s.avg_daily_sales_quantity = m.max_avg_daily_sales_quantity
    ORDER BY s.product_id
""").fetchdf()
solution
```

Observation-5 Determining the top 3 products with the highest average total sales quantity  for each month

```python
soln = con.sql("""
    WITH ranked_sales AS (
        SELECT product_id,Month,
               SUM(quantity_sold) AS total_sales,
               RANK() OVER (PARTITION BY Month 
                            ORDER BY SUM(quantity_sold) DESC) AS rank
        FROM updated_sales_table
        GROUP BY product_id,Month
    )
    SELECT Month,product_id,
           total_sales,
    FROM ranked_sales
    WHERE rank <= 3
    ORDER BY Month, rank
""").fetchdf()

soln
```

Observtion-6 Computing the month-to-month percentage change in averge daily sales quantiy for each product

```python
sol = con.sql("""
    WITH 
    ranked_sales AS (
        SELECT product_id,
               Month,
               avg_daily_sales_quantity,
               LAG(avg_daily_sales_quantity) OVER (PARTITION BY product_id ORDER BY Month) AS prev_avg_daily_sales_quantity
        FROM avg_sales
    )
    SELECT product_id,
           Month,
           avg_daily_sales_quantity,
           prev_avg_daily_sales_quantity,
           CASE
               WHEN prev_avg_daily_sales_quantity IS NULL THEN NULL
               ELSE ((avg_daily_sales_quantity - prev_avg_daily_sales_quantity) / prev_avg_daily_sales_quantity) * 100
           END AS percentage_change
    FROM ranked_sales
    ORDER BY product_id, Month
""").fetchdf()

sol
```
